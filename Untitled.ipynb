```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from meteostat import Daily, Point
from datetime import datetime, timedelta
from scipy.stats import zscore


cities = {
    "Tashkent": Point(41.2995, 69.2401),
    "New York": Point(40.7128, -74.0060),
    "London": Point(51.5074, -0.1278),
    "Tokyo": Point(35.6895, 139.6917),
    "Moscow": Point(55.7558, 37.6173)
}


end = datetime.today()
start = end - timedelta(days=5 * 365)


fig, axes = plt.subplots(len(cities), 1, figsize=(10, 5 * len(cities)))

for i, (city, location) in enumerate(cities.items()):
    data = Daily(location, start, end)
    data = data.fetch()
    
    df = data[['tavg']].dropna()
    
    df['z_score'] = zscore(df['tavg'])
    df['anomaly'] = df['z_score'].abs() > 2  
    ax = axes[i]
    ax.scatter(df.index, df['tavg'], c=df['anomaly'].map({True: 'red', False: 'blue'}), s=5)
    ax.set_title(f"{city} - So‘nggi 5 yillik harorat")
    ax.set_xlabel("Sana")
    ax.set_ylabel("O'rtacha harorat (°C)")

plt.tight_layout()
plt.show()
```
